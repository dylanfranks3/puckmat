```python
import pandas as pd
from statistics import mean
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df = pd.read_csv("data.csv")

df['deviceTimestamp'] = pd.to_datetime(df['deviceTimestamp'])

originalValues = []
for j,item in df.iterrows(): # adding the original data to the array
    originalValues.append(item.values[1]) 

for globalWindow in range(1,10):
    for threshold in range(0,32000,2000):
        for localWindow in range(1,10):
            newValues = [None for i in range(globalWindow+localWindow)]
            for idx,i in enumerate(originalValues):
                if idx >= globalWindow + localWindow and idx <= (len(originalValues) - globalWindow -localWindow- 1):
                    leftGlobalWindow = originalValues[idx-globalWindow-localWindow:idx-localWindow]
                    rightGlobalWindow = originalValues[idx+1+localWindow:idx+globalWindow+1+localWindow]
                    globalWindowArr = leftGlobalWindow + rightGlobalWindow

                    firstCondition = False
                    secondCondition = False
                    # first condition
                    for globalWindowItem in globalWindowArr: 
                        if abs(globalWindowItem - i) > threshold:
                            firstCondition = True
                    # second condition
                    if abs(mean(rightGlobalWindow) - mean(leftGlobalWindow)) < threshold/4:
                        secondCondition = True

                    if firstCondition and secondCondition:
                        newValues.append(None)
                    else:
                        newValues.append(i)
                    changed = False

            newValues = newValues + [None for i in range(globalWindow+localWindow)]
            print (len(newValues),len(originalValues))
            newData = pd.Series(newValues)
            df['globalWindowedColumn'] = newData

            plt.figure(figsize=(30, 10))
            plt.plot(df["deviceTimestamp"], df["globalWindowedColumn"])
            plt.subplots_adjust(bottom=0.19)

            # Labels and title
            plt.xlabel("Date",fontsize=20)
            plt.ylabel("Reading",fontsize=20)
            plt.title(f"Puck Mat Readings, Threshold={threshold}, Global Window={globalWindow}, Local Window={localWindow}",fontsize=20)

            # Rotate x-ticks
            plt.xticks(rotation=45, fontsize=14)

            # Y-axis limit
            plt.ylim(0, 70000)
            plt.yticks(fontsize=20)

            # Set x-ticks every half minute
            locator = mdates.HourLocator(interval=3)  # Set ticks at 30-second intervals
            formatter = mdates.DateFormatter("%-d-%a %H:%M:%S")  # Format ticks as HH:MM:SS
            plt.gca().xaxis.set_major_locator(locator)
            plt.gca().xaxis.set_major_formatter(formatter)

            # Save figure
            print (f"imgs/{threshold}_{globalWindow}_{localWindow}.jpeg")
            plt.savefig(f"imgs/{threshold}_{globalWindow}_{localWindow}.jpeg", dpi=500)
            plt.close()



""" for img in *.jpg
    ffmpeg -i $img -vf "crop=12500:4631:1100:350" $img
end """

```
